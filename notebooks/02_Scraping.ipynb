```python
valoraciones = []
driver = webdriver.Chrome()
driver.get("https://minimalismbrand.com/collections/sudaderas")
driver.maximize_window()
driver.execute_script("window.scrollBy(0, 750);")
producto1 = driver.find_element("xpath","/html/body/main/div/div[5]/div/div[2]/div[3]/div[2]/div/div[4]").click()
sleep(1)
driver.find_element("xpath","/html/body/main/div/div[1]/div/div[2]/div[2]/div/span").click()
valoracion = driver.find_element("xpath","/html/body/main/div/div[7]/div/div/div/div/div/div[2]/div/div[1]/span[1]/span").text
temp_df = pd.DataFrame(columns=["columna"])
temp_df["columna"] = [valoracion]*5
val = temp_df["columna"].tolist()
for v in val: 
    valoraciones.append(v)
driver.back()
producto2 = driver.find_element("xpath","/html/body/main/div/div[5]/div/div[2]/div[3]/div[2]/div/div[6]").click()
sleep(1)
valoracion = driver.find_element("xpath","/html/body/main/div/div[7]/div/div/div/div/div/div[2]/div/div[1]/span[1]/span").text
temp_df = pd.DataFrame(columns=["columna"])
temp_df["columna"] = [valoracion]*5
val = temp_df["columna"].tolist()
for v in val: 
    valoraciones.append(v)
driver.quit()
driver = webdriver.Chrome()
driver.get("https://minimalismbrand.com/collections/sweater-men")
```
